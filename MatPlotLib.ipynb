```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
#x
```

```python
#y
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('TITLE')
```

```python
plt.subplot(121)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b--')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])#[X0,Y0,Width,Height of the graph]
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('OOmethod')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(5,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x Square')
axes.plot(x,x**3,label='x Cubed')
axes.legend()
```
